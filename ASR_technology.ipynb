```python
# ! pip install git+https://github.com/huggingface/transformers -q

import os
import torch
import torchaudio
from transformers import pipeline
import librosa
from evaluate import load

from torchmetrics.text import WordErrorRate
from torchmetrics.text import CharErrorRate

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

whisper = pipeline('automatic-speech-recognition', model = 'openai/whisper-medium')

source_dir = "/content"
count = 0
char_err_count = 0
word_err_count = 0


def text_define(speech_in):
  speech = whisper(speech_in)
  print(speech)
  return speech

def text_expected(file_select):
  with open(f"{source_dir}/{file_select}.txt") as f:
    lines = f.readlines()
    print(lines)
  return lines

def err_rate(pred, expect):
  word_err = WordErrorRate(pred, expect)
  char_err = CharErrorRate(pred, expect)
  return word_err, char_err

entries = os.listdir(source_dir)
for file_list in entries:
  if file_list.endswith(".wav"):

    speech, fs = librosa.load(f"{source_dir}/{file_list}")
    predicted = text_define(speech)
    expected = text_expected(f'{file_list[0:-8]}')

    count = count+1

    char_word = load("cer")
    word_word = load("wer")

    cer_score = char_word.compute(predictions=predicted, references=expected)
    wer_score = word_word.compute(predictions=predicted, references=expected)

    #wer_score, cer_score = err_rate(predicted, expected)
    print(f"CER= {cer_score} : : WER =  {wer_score}")

    char_err_count = char_err_count + cer_score
    word_err_count = word_err_count + wer_score

    print(f"wer = {word_err_count} :: cer = {char_err_count} :: count = {count}")
  print(f"Word_error_rate = {word_err_count/count} Character_error_rate = {char_err_count/count}")
```

```python
# ! pip install git+https://github.com/huggingface/transformers -q

import os
import torch
import torchaudio
from transformers import pipeline
import librosa
from jiwer import cer
from jiwer import wer


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

whisper = pipeline('automatic-speech-recognition', model = 'openai/whisper-medium')

source_dir = "/content/drive/MyDrive/Distance_ASR/D1/Hindi"
count = 0
char_err_count = 0
word_err_count = 0


def text_define(speech_in):
  transcribe = pipeline(task="automatic-speech-recognition", model="vasista22/whisper-hindi-medium", chunk_length_s=30, device=device)#.to(device)
  transcribe.model.config.forced_decoder_ids = transcribe.tokenizer.get_decoder_prompt_ids(language="hi", task="transcribe")#.to(device)
  print('Transcription: ', transcribe(speech_in)["text"])
  speech = transcribe(audio)["text"]

  return speech

def text_expected(file_select):
  with open(f"{source_dir}/{file_select}.txt") as f:
    lines = f.readlines()
    print(lines)
  return lines

def err_rate(pred, expect):
  word_err = WordErrorRate(pred, expect)
  char_err = CharErrorRate(pred, expect)
  return word_err, char_err

entries = os.listdir(source_dir)
for file_list in entries:
  if file_list.endswith(".wav"):

    speech, fs = librosa.load(f"{source_dir}/{file_list}")
    predicted = text_define(speech)
    expected = text_expected(f'{file_list[0:-4]}')

    count = count+1

    cer_score = cer(predicted, expected)
    wer_score = wer(predicted, expected)

    print(f"CER= {cer_score} : : WER =  {wer_score}")

    char_err_count = char_err_count + cer_score
    word_err_count = word_err_count + wer_score

    print(f"wer = {word_err_count} :: cer = {char_err_count} :: count = {count}")
  print(f"Word_error_rate = {word_err_count/count} Character_error_rate = {char_err_count/count}")
```

```python
import os
import numpy as np
import torch
import torchaudio
import torchmetrics
from torch import tensor
from torchmetrics.functional.audio import signal_noise_ratio
from torchmetrics.functional.audio import signal_distortion_ratio

#si_snr = ScaleInvariantSignalNoiseRatio()
#si_sdr = ScaleInvariantSignalDistortionRatio()

#source_dir1 = '/content/drive/MyDrive/Mic Array recordings/AIRA Corpus/Anechoic Chamber Recordings/1 Source/45__3000'
#dest_dir1 = '/content/drive/MyDrive/Mic Array recordings/AIRA Corpus/Anechoic Chamber Recordings/1 Source/45__3000'

source_dir1 = '/content/drive/MyDrive/Mic Array recordings/AIRA Corpus/Office A Recordings/1 Source/0_0000'
#dest_dir1 = '/content/drive/MyDrive/Mic Array recordings/AIRA Corpus/Office A Recordings/1 Source/0_0000'

sampling_rate = 48000
count = 0
SNR = 0
SDR = 0

for x in range(10):
  entries = os.listdir(f"{source_dir1}/Evaluation_{x+1}")
  for file1 in entries:
    if file1.endswith('MVcos.wav'):

      #Print the file name:
      #Load the beamformed variable as predicted

      predicted, fs_pred = torchaudio.load(f"{source_dir1}/Evaluation_{x+1}/{file1}")
      predicted = torch.FloatTensor(predicted[0, 0:1300000])

      #print(predicted)

      #Load the bluetooth signal (Ground truth) as the expected or the target
      target, fs_targ = torchaudio.load(f"{source_dir1}/Evaluation_{x+1}/pristine_channel1.wav")
      target = torch.FloatTensor(target[0, 0:1300000])

      #print(target)

      snr = signal_noise_ratio(predicted, target)
      sdr = signal_distortion_ratio(predicted, target)

      SNR = SNR + snr
      SDR = SDR + sdr

      count = count+1

      print(f'The present SNR: {SNR} for the count : {count}')
      print(f'The present SDR: {SDR} for the count : {count}')


```
